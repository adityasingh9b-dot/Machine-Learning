```python
import pandas as pd
import numpy as np

# Set seed for reproducibility
np.random.seed(42)

# Number of samples
n = 1000

# Generate synthetic data
initial_payment = np.random.randint(5000, 100000, size=n)
last_payment = np.random.randint(1000, 50000, size=n)
credit_score = np.random.randint(300, 851, size=n)
house_no = np.random.randint(100, 10000, size=n)

# Generate 'Result' based on some logic (you can change logic later)
# For now, let's say higher credit score + higher payments = higher chance of repayment
result = ((credit_score > 600) & (initial_payment > 30000) & (last_payment > 10000)).astype(int)

# Create DataFrame
df = pd.DataFrame({
    "Result": result,
    "Initial Payment": initial_payment,
    "Last Payment": last_payment,
    "Credit Score": credit_score,
    "House No.": house_no
})

# Show first few rows
print(df.head())

# Save to CSV
df.to_csv("loan_prediction_data.csv", index=False)
```

```python
data = pd.read_csv("loan_prediction_data.csv")
data.head()
```

```python
print("Length (Rows) of Data:", len(data))
print("Shape (Dimension) of Data:", data.shape)
```

```python
X = data.values[0:1000, 1:5]
y = data.values[0:1000, 0]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

clf_entropy = DecisionTreeClassifier(criterion='entropy', random_state=0, max_depth=3, min_samples_leaf=5)
clf_entropy.fit(X_train, y_train)
```

```python
y_pred_en = clf_entropy.predict(X_test)
print(y_pred_en)
print(y_test)
```

```python
print(accuracy_score(y_pred_en,y_test) * 100, "%" )
```
